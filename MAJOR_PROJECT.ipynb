```
df_new = df[['Date','active_cases']]
df_new.info()
```

```
df_new = df_new.rename(columns={'Date':'ds','active_cases':'y'})
df_new
```

```
import pandas as pd
from fbprophet import Prophet
```

```
m = Prophet()
m.fit(df_new)
```

```
future = m.make_future_dataframe(periods=365)
future.tail()
```

```
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```
fig1 = m.plot(forecast)
```

```
fig2 = m.plot_components(forecast)
```

```
from fbprophet.plot import plot_plotly, plot_components_plotly

plot_plotly(m, forecast)
```

```
plot_components_plotly(m, forecast)
```

**highest active cases in which state.?**

**Highest Cured State.?** bold text

```
max=df['Cured'].max()
max_df=df[['State/UnionTerritory','Cured']][df['Cured']>max-1]
max_df
```

**Highest Death State.?**

```
max=df['Deaths'].max()
max_df=df[['State/UnionTerritory','Deaths']][df['Deaths']>max-1]
max_df
```

**Least Affected State.?**

```
min=df['Confirmed'].min()
min_df=df[['State/UnionTerritory','Confirmed']][df['Confirmed']<min+1]
min_df
```
